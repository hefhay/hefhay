```python
#to remove cases where confirmed cases are less than 10
covid = covid[~(covid.Confirmed <10)]
covid.head(20)
```

TO GET THE REGION WHERE MAXIMUM NUMBER OF CASES WERE RECORDEED

```python
covid.groupby('Region').Confirmed.sum().sort_values(ascending = False).head(20)
```

```python
covid.groupby('Region').Deaths.sum().sort_values(ascending  = True).head(20)
```
